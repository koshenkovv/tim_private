```python
df = pd.read_csv('dfMoscow4(demo).csv')
df
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.drop(['properties.id'], axis = 1)
df = df.drop(['properties.point.lat'], axis = 1)
df = df.drop(['properties.point.long'], axis = 1)
df = df.drop(['properties.category'], axis = 1)
df = df.drop(['properties.severity'], axis = 1)
df = df.drop(['properties.dead_count'], axis = 1)
df = df.drop(['properties.parent_region'], axis = 1)
df = df.drop(['properties.participants_count'], axis = 1)
df = df.drop(['Участие пешеходов'], axis = 1)
df = df.drop(['Водитель'], axis = 1)
df
```

```python
df.drop_duplicates(subset = 'properties.address', inplace = True)
```
