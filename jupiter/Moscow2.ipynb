# Предобработка данных

Загрузка необходимых библиотек

```python
import pandas as pd
pd.options.display.max_columns = 1000
```

Открытие датасета

```python
df = pd.read_csv('dfMoscow.csv')
df
```

```python
df.loc[0]['properties.vehicles']
```

Получим информацию о столбцах таблицы

```python
df.info()
```

Удаление неинформативных признаков

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.drop(['type'], axis = 1)
df = df.drop(['geometry.type'], axis = 1)
df = df.drop(['properties.tags'], axis = 1)
df = df.drop(['properties.nearby'], axis = 1)
df = df.drop(['properties.scheme'], axis = 1)
df = df.drop(['properties.participants'], axis = 1)
df = df.drop(['properties.datetime'], axis = 1)
df
```

Проверяем дубликаты так как id ДТП не может быть одинаковым

```python
df.duplicated(subset='properties.id', keep='first').sum()
```

Добавим новый столбец, где узнаем участвовали ли пешеходы в ДТП

```python
df['properties.participant_categories'].apply(type)
content = []
for i in range(0, len(df['properties.participant_categories'])):
    if 'Пешеходы' in df.loc[i]['properties.participant_categories']:
        content.append(1)
    else:
        content.append(0)
        
df["Участие пешеходов"] = content
```

```python
df
```

Удаляем столбец participant_categories, он нам больше не интересен

```python
df = df.drop(['properties.participant_categories'], axis = 1)
```

Добавим новый столбец, где узнаем какого пола был водитель

```python
content = []
for i in range(0, len(df['properties.vehicles'])):
    if "'role': 'Водитель', 'gender': 'Мужской'" in df.loc[i]['properties.vehicles']:
        content.append('Мужчина')
    elif "'role': 'Водитель', 'gender': 'Женский'" in df.loc[i]['properties.vehicles']:
        content.append('Женщина')
    else:
        content.append('Неизвестно')
df["Водитель"] = content
```

```python
df_sex = df.groupby(['Водитель']).size().reset_index(name='Кол-во ДТП')
df_sex
```

```python
df.loc[0]['properties.vehicles']
```

Удаляем столбец properties.vehicles, он нам больше не интересен

```python
df = df.drop(['properties.vehicles'], axis = 1)
```

```python
df
```

```python
df_light = df.groupby(['properties.light']).size().reset_index(name='Кол-во ДТП')
df_light.sort_values(by='Кол-во ДТП')
```

Почистим колонку properties.light, определим 5 видов

```python
content = []
for i in range(0, len(df['properties.light'])):
    if "В темное время суток, освещение не включено" in df.loc[i]['properties.light']:
        content.append('В темное время суток, освещения нет')
    elif "В темное время суток, освещение отсутствует" in df.loc[i]['properties.light']:
        content.append('В темное время суток, освещения нет')
    elif "В темное время суток, освещение включено" in df.loc[i]['properties.light']:
        content.append('В темное время суток, освещения есть')
    elif "Светлое время суток" in df.loc[i]['properties.light']:
        content.append('Светлое время суток')
    else:
        content.append('Неизвестно')
df["properties.light"] = content
```

Сделаем проверку

```python
df_light = df.groupby(['properties.light']).size().reset_index(name='Кол-во ДТП')
df_light.sort_values(by='Кол-во ДТП')
```

Почистим колонку properties.weather, определим 5 видов

```python
df_weather = df.groupby(['properties.weather']).size().reset_index(name='Кол-во ДТП')
df_weather.sort_values(by='Кол-во ДТП')
```

```python
for i in range(0, len(df['properties.weather'])):
    if "Пасмурно" in df.loc[i]['properties.weather']:
        df.at[i, 'properties.weather'] = 'Пасмурно'
    elif "Ясно" in df.loc[i]['properties.weather']:
        df.at[i, 'properties.weather'] = 'Ясно'
    elif "Дождь" in df.loc[i]['properties.weather']:
        df.at[i, 'properties.weather'] = 'Дождь'
    elif "Снегопад" in df.loc[i]['properties.weather']:
        df.at[i, 'properties.weather'] = 'Снегопад'
    else:
        df.at[i, 'properties.weather'] = 'Неизвестно'
```

Сделаем проверку

```python
df_weather = df.groupby(['properties.weather']).size().reset_index(name='Кол-во ДТП')
df_weather.sort_values(by='Кол-во ДТП')
```

Почистим колонку properties.road_conditions, определим 3 вида

```python
df_road = df.groupby(['properties.road_conditions']).size().reset_index(name='Кол-во ДТП')
df_road.sort_values(by='Кол-во ДТП').tail(10)
```

```python
for i in range(0, len(df['properties.road_conditions'])):
    if "Мокрое" in df.loc[i]['properties.road_conditions']:
        df.at[i, 'properties.road_conditions'] = 'Мокрое'
    elif "Сухое" in df.loc[i]['properties.road_conditions']:
        df.at[i, 'properties.road_conditions'] = 'Сухое'
    else:
        df.at[i, 'properties.road_conditions'] = 'Другое'
```

Сделаем проверку

```python
df_weather = df.groupby(['properties.road_conditions']).size().reset_index(name='Кол-во ДТП')
df_weather.sort_values(by='Кол-во ДТП')
```
