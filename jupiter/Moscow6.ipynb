Загрузка необходимых библиотек

```python
import pandas as pd
pd.options.display.max_columns = 1000
```

Открытие датасета

```python
df = pd.read_csv('dfMoscow5.csv')
df = df.drop(['Unnamed: 0'], axis = 1)
```

Проверим, не ли одинаковых дорожных участков

```python
df['Адрес'].nunique()
```

Добавим новую колонку, которая будет отражать среднее значение тяжести

```python
def severity(df):
    s1, s2, s3 = df['Легкий'], df['Тяжёлый'], df['С погибшими']
    i = (1*s1+2*s2+3*s3)/(s1+s2+s3)
    return i
```

```python
df['Среднее значение тяжести'] = df.apply(lambda x: severity(x), axis = 1)
```

```python
df
```

```python
df['Среднее значение тяжести'].describe()
```

Добавим новую колонку, которая будет отражать индекс опасности дорожного участка

```python
def index(df):
    k, s = df['Кол-во ДТП по данному адресу'], df['Среднее значение тяжести']
    return int(k*s)
```

```python
df['Индекс'] = df.apply(lambda x: index(x), axis = 1)
```
