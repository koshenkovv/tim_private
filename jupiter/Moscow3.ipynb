# Добавление новых аттребутов

Загрузка необходимых библиотек

```python
import pandas as pd
pd.options.display.max_columns = 1000
```

Открытие датасета

```python
df = pd.read_csv('dfMoscow2.csv')
df
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df = df.drop(['properties.weather'], axis = 1)
df = df.drop(['properties.light'], axis = 1)
df = df.drop(['properties.road_conditions'], axis = 1)
df = df.drop(['properties.injured_count'], axis = 1)
df = df.drop(['geometry.coordinates'], axis = 1)
```

```python
df['properties.address'].nunique()
```

```python
df['properties.address'].isnull().sum()
```

```python
df = df.fillna('Неизвестно')
```

```python
df['properties.address'].isnull().sum()
```

```python
df
```

```python
df2 = df.groupby(['properties.category']).size().reset_index(name='Кол-во ДТП')
df2.sort_values(by='Кол-во ДТП')
```

```python
category = ['Иной вид ДТП', 'Съезд с дороги', 'Опрокидывание', 'Наезд на велосипедиста', 'Падение пассажира', 'Наезд на стоящее ТС', 'Наезд на препятствие', 'Наезд на пешехода', 'Столкновение']
for i in range(0, len(df['properties.category'])):
    if df.loc[i]['properties.category'] not in category:
        df.at[i, 'properties.category'] = 'Иной вид ДТП'
```

Проверяем

```python
df2 = df.groupby(['properties.category']).size().reset_index(name='Кол-во ДТП')
df2.sort_values(by='Кол-во ДТП')
```

Воспользуемся бинаризацией

```python
df2 = pd.get_dummies(df['properties.category'])
df2
```

```python
df = df.join(df2)
```

```python
df
```

```python
df3 = pd.get_dummies(df['properties.severity'])
df3
```

```python
df = df.join(df3)
df
```

Данные добавлены, сохраним датасет

```python
df.to_csv('dfMoscow3.csv', index=False)
```
